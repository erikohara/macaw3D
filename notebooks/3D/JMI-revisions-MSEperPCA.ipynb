```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = '/work/forkert_lab/erik/JMI_paper_2/revisions'
```

```python
pca3d = pd.read_csv(path + "/PCA3D_MSE_per_nevecs.csv",low_memory=False)
pca2d_axial = pd.read_csv(path + "/PCA_2Daxial_MSE_per_nevecs.csv",low_memory=False)
pca2d_coronal = pd.read_csv(path + "/PCA_2Dcoronal_MSE_per_nevecs.csv",low_memory=False)
pca2d_sagittal = pd.read_csv(path + "/PCA_2Dsagittal_MSE_per_nevecs.csv",low_memory=False)
```

```python
pca3d.plot(x='nevecs', y='MSEloss')
plt.title('3D PCA')
plt.xlabel('Number of Principal Components')
plt.ylabel('Mean Square Error')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6)) 
# Plot each dataframe
pca2d_axial.plot(x='nevecs', y='MSEloss', ax=ax, label='Axial slices')
pca2d_coronal.plot(x='nevecs', y='MSEloss', ax=ax, label='Coronal slices')
pca2d_sagittal.plot(x='nevecs', y='MSEloss', ax=ax, label='Sagittal slices')

plt.title('2D PCA')
plt.xlabel('Number of Principal Components')
plt.ylabel('Mean Square Error')
plt.legend()  # Show the labels for each line
plt.show()
```
