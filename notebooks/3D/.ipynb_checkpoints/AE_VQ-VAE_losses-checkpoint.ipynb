```python
import pickle
import matplotlib.pyplot as plt
```

```python
ae_path_2000 = '/work/forkert_lab/erik/MACAW/models/AE3D_2000_previous'
ae_path_4000 = '/work/forkert_lab/erik/MACAW/models/AE3D_4000_previous'
ae_path_6000 = '/work/forkert_lab/erik/MACAW/models/AE3D_6000_previous'
ae_path_8000 = '/work/forkert_lab/erik/MACAW/models/AE3D_8000_previous'
ae_path_10000 = '/work/forkert_lab/erik/MACAW/models/AE3D_10000_previous'
vqvae_path = '/work/forkert_lab/erik/MACAW/models/vqvae3D_64'
vqvae_path_32 = '/work/forkert_lab/erik/MACAW/models/vqvae3D_32'
vqvae_path_16 = '/work/forkert_lab/erik/MACAW/models/vqvae3D_16'
vqvae_path_8 = '/work/forkert_lab/erik/MACAW/models/vqvae3D_8'
vqvae_path_4 = '/work/forkert_lab/erik/MACAW/models/vqvae3D_4'
```

# Autoencoder

```python
with open(ae_path_2000 + '/losses.pkl','rb') as f1:  
    ae_losses_2000 = pickle.load(f1)
with open(ae_path_4000 + '/losses.pkl','rb') as f2:  
    ae_losses_4000 = pickle.load(f2)
with open(ae_path_6000 + '/losses.pkl','rb') as f3:  
    ae_losses_6000 = pickle.load(f3)
with open(ae_path_8000 + '/losses.pkl','rb') as f4:  
    ae_losses_8000 = pickle.load(f4)
with open(ae_path_10000 + '/losses.pkl','rb') as f5:  
    ae_losses_10000 = pickle.load(f5)
```

```python
epochs_2000 = range(1, len(ae_losses_2000['val_loss'])+1)
epochs_4000 = range(1, len(ae_losses_4000['val_loss'])+1)
epochs_6000 = range(1, len(ae_losses_6000['val_loss'])+1)
epochs_8000 = range(1, len(ae_losses_8000['val_loss'])+1)
epochs_10000 = range(1, len(ae_losses_10000['val_loss'])+1)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(epochs_2000, ae_losses_2000['train_loss'], label='2000 - training loss')
plt.plot(epochs_2000, ae_losses_2000['val_loss'], label='2000 - Validation loss')
plt.plot(epochs_4000, ae_losses_4000['train_loss'], label='4000- Training Loss')
plt.plot(epochs_4000, ae_losses_4000['val_loss'], label='4000 - Validation Loss')
plt.plot(epochs_6000, ae_losses_6000['train_loss'], label='6000 - Training Loss')
plt.plot(epochs_6000, ae_losses_6000['val_loss'], label='6000 - Validation Loss')
plt.plot(epochs_8000, ae_losses_8000['train_loss'], label='8000 - Training Loss')
plt.plot(epochs_8000, ae_losses_8000['val_loss'], label='8000 - Validation Loss')
plt.plot(epochs_10000, ae_losses_10000['train_loss'], label='10000 - Training Loss')
plt.plot(epochs_10000, ae_losses_10000['val_loss'], label='10000 - Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('torch.nn.MSELoss')
plt.title('AE - Training and Validation Loss Curves')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print("Traning lost")
print(min(ae_losses_2000['train_loss']))
print(min(ae_losses_4000['train_loss']))
print(min(ae_losses_6000['train_loss']))
print(min(ae_losses_8000['train_loss']))
print(min(ae_losses_10000['train_loss']))

#Validation lost
print("Validation lost")
print(min(ae_losses_2000['val_loss']))
print(min(ae_losses_4000['val_loss']))
print(min(ae_losses_6000['val_loss']))
print(min(ae_losses_8000['val_loss']))
print(min(ae_losses_10000['val_loss']))
```
