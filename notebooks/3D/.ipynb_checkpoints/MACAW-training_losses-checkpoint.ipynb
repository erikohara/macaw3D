# PD - PCA 3D with 80 ncomps

```python
with open(pd_pca3D_80ncomps + '/losses.pkl','rb') as fpca:  
    pd_80ncomps_losses = pickle.load(fpca)
```

```python
len(pd_80ncomps_losses.items())
```

```python
# Set up a grid with 3 rows and 4 columns
fig, axs = plt.subplots(3, 4, figsize=(15, 10))
plot_limit = 12

# Flatten the axes array for easy iteration
axs = axs.flatten()

# Plot each loss value in the grid
nplot = 0
for i,key in enumerate(pd_80ncomps_losses.keys()):
    if nplot < plot_limit:
        nplot += 1
    else:
        break
    axs[i].plot(np.array(pd_80ncomps_losses[key][0]), 'r+')
    axs[i].plot(np.array(pd_80ncomps_losses[key][1]), 'b.')
    axs[i].set_ylim(bottom=-500,top=10000)
    axs[i].set_title(f'Plot {key}')

plt.tight_layout()
plt.show()
```

```python
# Set up a grid with 3 rows and 4 columns
fig, axs = plt.subplots(5, 4, figsize=(15, 10))
plot_ini = 20
plot_limit = 40

# Flatten the axes array for easy iteration
axs = axs.flatten()

# Plot each loss value in the grid
for i,key in enumerate(ae_losses.keys()):
    if i < plot_limit:
        nplot += 1
    else:
        break
    if i >= plot_ini:
        axs[i-plot_ini].plot(np.array(ae_losses[key][0]), 'r+')
        axs[i-plot_ini].plot(np.array(ae_losses[key][1]), 'b.')
        axs[i-plot_ini].set_title(f'Plot {key}')

plt.tight_layout()
plt.show()
```
