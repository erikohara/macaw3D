```python
import pandas as pd
import numpy as np
import nibabel as nib
import os
import sys
from pathlib import Path
sys.path.append(str(Path(os.getcwd()).parent.parent) +'/')
from utils.visualize import img_grid
import ipywidgets as ipyw
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
```

```python
generated_path = '/work/forkert_lab/erik/MACAW/cf_images/PD_PCA3ind_age_sex_indsite_whole_obj_cosine_gpu'
original_folder = '/work/forkert_lab/mitacs_dataset/affine_using_nifty_reg'
data_path = '/home/erik.ohara/SFCN_PD_scanner'
nsamples = 5
z_initial = 0
z_fim = 159
```

# Difference map

```python
df = pd.read_csv(data_path + '/all_df_2.csv',low_memory=False)
```

```python
rands = np.random.randint(0,len(df),nsamples) 
```

```python
Subjects = df['Subject'][rands].to_list()
Subjects
```

```python
sites = df['Study'][rands].to_list()
sites
```

```python
images_original = []
images_generated = []
for participant_id in Subjects:
    print(participant_id)
    image_load = nib.load(os.path.join(generated_path, participant_id + '.nii.gz')).get_fdata()
    images_generated.append(image_load)
    image_load_original = nib.load(os.path.join(original_folder, participant_id + ".nii.gz")).get_fdata()
    images_original.append(image_load_original)
```

```python
print(images_original[1].max())
print(images_original[1].min())
print()
print(images_generated[1].max())
print(images_generated[1].min())
```

```python
diff = [np.subtract(a,s) for a,s in zip(images_original,images_generated)]
```

```python
def plot_slice(z_slice):
    order_slice = z_slice - z_initial
    plt.rcParams["figure.figsize"] = (20,5)
    img_grid([one_image[:,:,order_slice] for one_image in images_original],cols=nsamples,titles=sites)
    img_grid([one_image[:,:,order_slice] for one_image in images_generated],cols=nsamples)
    img_grid([one_image[:,:,order_slice] for one_image in diff],cols=nsamples, cmap='seismic')
```

```python
ipyw.interact(plot_slice, 
              z_slice=ipyw.IntSlider(min=z_initial, max=z_fim, step=1, continuous_update=False, description='Image Slice:'))
```

```python
fig, axs = plt.subplots(5, 2, figsize=(15, 10))

# Flatten the axes array for easy iteration
for i in range(len(images_generated)):
    axs[i,0].hist(images_original[i].reshape(-1), bins='auto') 
    axs[i,0].set_xlim(left=-.9, right=1)
    axs[i,0].set_ylim(top=5e4)
    axs[i,0].set_title(f"Histogram of {Subjects[i]} from {sites[i]}")
    
    axs[i,1].hist(images_generated[i].reshape(-1), bins='auto') 
    axs[i,1].set_xlim(left=-.9, right=1)
    axs[i,1].set_ylim(top=5e4)
    axs[i,1].set_title(f"Histogram of {Subjects[i]} as UKBB")
    
plt.tight_layout()
plt.show()
```

# Original images histogram per site

```python
sites = df['Study'].unique()
```

```python
df_persite = {}
subjects_persite = {}
for site in sites:
    print(site)
    df_site = df[df['Study'] == site].reset_index()
    df_persite[site] = df_site
    rands = np.random.randint(0,len(df_site),nsamples) 
    subjects_persite[site] = df_site['Subject'][rands].to_list()
```

```python
images_original_persite = {} 
for site in tqdm(sites):
    images_original_persite[site] = []
    for participant_id in subjects_persite[site]:
        image_load_original = nib.load(os.path.join(original_folder, participant_id + ".nii.gz")).get_fdata()
        images_original_persite[site].append(image_load_original)
```

```python
fig, axs = plt.subplots(10, 5, figsize=(15, 10))

# Flatten the axes array for easy iteration
for idx,site in enumerate(sites):
    for sample in range(nsamples):
        axs[idx,sample].hist(images_original_persite[site][sample].reshape(-1), bins='auto') 
        axs[idx,sample].set_xlim(left=-.9, right=1)
        axs[idx,sample].set_ylim(top=5e4)
        axs[idx,sample].set_title(f"{site}")
    
plt.tight_layout()
plt.show()
```
